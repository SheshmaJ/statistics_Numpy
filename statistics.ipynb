```python
score=[23,24,25,32,45,12,43,67,45,32,56,32,200]
```

```python
q1,q3=np.percentile(score,[25,75])
print(q1,q3)
```

```python
IQR=q3-q1
lower_fence=q1-1.5*(IQR)
upper_fence=q3+1.5*(IQR)
print(lower_fence,upper_fence)
```

```python
import numpy as np
```

```python
s=np.random.normal(0.5,0.2,1000)                             #loc=o.5 means mean or center,0.2 is standard deviation and 1000 is sample size
s
```

```python
sns.histplot(s,kde=True)
```

```python
#log normmal distribution
```

```python
mu,sigma=3,1
s=np.random.lognormal(mu,sigma,100)
```

```python
sns.histplot(s,kde=True)
```

```python
sns.histplot(np.log(s),kde=True)
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
stats.probplot(np.log(s), dist="norm", plot=plt)             #checking if log distribution is normal using qq plt
plt.show()
```

```python
stats.probplot(s, dist="norm", plot=plt)                                        #checking if distribution is normal using qq plt
plt.show()
```
